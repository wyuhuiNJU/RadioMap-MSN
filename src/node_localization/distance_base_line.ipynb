```python
# 分类网络有个基线, 比如4分类网络, 基线是1/4, 即随预测的准确度, 准确度>1/4代表有分类效果
# 回归网络的基线, 应该是随机预测的值与实际值的误差平均值
# 此程序模拟随机取点, 预测随机值

m = 50
n = 80
times = 1e5
```

```python
import random
from matplotlib import pyplot as plt


def randomPoint(m, n):
    x = random.random() * m
    y = random.random() * n
    return [x, y]

def cal_distance(point_a, point_b):
    [x1, y1] = point_a
    [x2, y2] = point_b
    distance = ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5
    return distance

m = 30
n = 50
times = int(1e6)
average_distance = []

all_distances = 0
for i in range(times):
    point_a = randomPoint(m ,n)
    point_b = randomPoint(m ,n)
    distance = cal_distance(point_a, point_b)
    all_distances += distance
    average = all_distances / (i + 1)
    average_distance.append(average)
    # print(f'average: {average}')

plt.plot(average_distance)
plt.show()
```

```python
import pandas as pd

positions = pd.read_csv('../../data/project_170/positions.csv', index_col=0).to_numpy()
```

```python
import random
from matplotlib import pyplot as plt


def cal_distance(point_a, point_b):
    [x1, y1] = point_a
    [x2, y2] = point_b
    distance = ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5
    return distance

times = int(1e6)
average_distance = []
length = len(positions)

all_distances = 0
for i in range(times):
    point_a = positions[random.randint(0, length-1)]
    point_b = positions[random.randint(0, length-1)]
    distance = cal_distance(point_a, point_b)
    all_distances += distance
    average = all_distances / (i + 1)
    average_distance.append(average)
    # print(f'average: {average}')

plt.plot(average_distance)
plt.show()
```
