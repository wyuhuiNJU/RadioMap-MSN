```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt


log_1 = pd.read_csv('../../model/project_209/model_2/log_1/batchSize=512_scaler=128_depth=7.csv', index_col=0)
loss_1 = log_1['loss'].to_numpy()
log_2 = pd.read_csv('../../model/project_209/model_2/log_2/batchSize=256_scaler=128_depth=7.csv', index_col=0)
loss_2 = log_2['loss'].to_numpy()
log_3 = pd.read_csv('../../model/project_209/model_2/log_3/batchSize=128_scaler=128_depth=7.csv', index_col=0)
loss_3 = log_3['loss'].to_numpy()
log_4 = pd.read_csv('../../model/project_209/model_2/log_4/batchSize=64_scaler=128_depth=7.csv', index_col=0)
loss_4 = log_4['loss'].to_numpy()
log_5 = pd.read_csv('../../model/project_209/model_2/log_5/batchSize=1024_scaler=128_depth=7.csv', index_col=0)
loss_5 = log_5['loss'].to_numpy()


k = 25

loss_1 = loss_1[k:]
loss_2 = loss_2[k:]
loss_3 = loss_3[k:]
loss_5 = loss_5[k:]
loss_4 = loss_4[k:]
```

```python
xs = range(k, 200)
plt.plot(xs, loss_4, label='batchSize=64')
plt.plot(xs, loss_3, label='batchSize=128')
plt.plot(xs, loss_2, label='batchSize=256')
plt.plot(xs, loss_1, label='batchSize=512')
plt.plot(xs, loss_5, label='batchSize=1024')


plt.xlabel('epochs')
plt.ylabel('val loss')

plt.legend()
```

```python
time_1 = log_1['time'].to_numpy()
time_2 = log_2['time'].to_numpy()
time_3 = log_3['time'].to_numpy()
time_4 = log_4['time'].to_numpy()


plt.figure()
xs = range(len(time_1))
plt.plot(xs, time_3, label='batchSize=128')
plt.plot(xs, time_2, label='batchSize=256')
plt.plot(xs, time_1, label='batchSize=512')
plt.plot(xs, time_4, label='batchSize=1024')
# plt.xlim(0, len(time_1)+110)
plt.xlabel('epochs')
plt.ylabel('time each epoch(s)')


plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
