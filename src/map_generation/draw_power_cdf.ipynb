```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv('../../data/project_complex_91/data.csv', index_col=0)
all_power = data['power'].tolist()
all_power.sort()
cdf = np.arange(1, len(all_power) + 1) / len(all_power)
plt.plot(all_power, cdf)
plt.show()
```
